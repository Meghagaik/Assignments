```python
import pandas as pd
sample = pd.read_csv("Q9_b.csv")

sample.head()
```

```python
#********Skewness************#

from scipy import stats
stats.skew(sample,bias=False)
```

```python
#*******************kurtosis*****************#


stats.kurtosis(sample,bias=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
plt.plot(sample.WT, sample.SP);
plt.xlabel('Weight');
plt.ylabel('Speed');
```
