```python
import pandas as pd
sample = pd.read_csv("Q9_a.csv")

sample.head()
```

```python
#********Skewness************#

from scipy import stats
stats.skew(sample,bias=False)
```

```python
#*******************kurtosis*****************#


stats.kurtosis(sample,bias=False)
```
