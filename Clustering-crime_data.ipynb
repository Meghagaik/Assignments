```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt 
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import seaborn as sn
```

```python
crime = pd.read_csv("crime_data.csv")
crime
```

```python
#Normalizing data
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(crime.iloc[:,1:])
df_norm
```

```python
#Create Dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='complete'))
```

```python
#Create Cluster
hc = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity = 'euclidean')
```

```python
#save cluster for chart
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
crime['h_clusterid']=hc.labels_
crime
```

```python
#K-Means Clustering
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
crime_1 = pd.read_csv("crime_data.csv")
```

```python
#Elbow Curve
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow Curve')
plt.xlabel('number of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#selecting 4 clusters from the aboove screen plot which is the optimum number of clusters
model=KMeans(n_clusters=4)
model.fit(df_norm)

model.labels_  #getting the labels of cluster assigned to each row
```

```python
md = pd.Series(model.labels_) #converting numpy array into pandas series object
crime_1['clust']= md #creating a new column and assigning it to new column
crime_1
```

```python
crime_1.iloc[:,1:7].groupby(crime.clust).mean()
```

```python
#DBSCAN Clustering
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("crime_data.csv")
print(df.head())
```

```python
print(df.info())
```

```python
df.drop([df.columns[0]],axis=1,inplace=True)
df
```

```python
array=df.values
array
```

```python
stscaler = StandardScaler().fit(array)
x = stscaler.transform(array)
x
```

```python
dbscan = DBSCAN(eps = 0.9, min_samples=4)
dbscan.fit(x)
```

```python
#Noisy samples are given the label -1
dbscan.labels_
```

```python
cl = pd.DataFrame(dbscan.labels_,columns=['clusters'])
cl
```

```python
pd.concat([df,cl],axis=1)
```
